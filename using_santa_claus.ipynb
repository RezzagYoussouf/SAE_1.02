```python
# tour_from_closest_city = construit un tour en ajoutant successivement la ville la plus proche .
# best_tour_from_closest_city = cet algorithme se répéte en essayant chaque ville comme ville de départ .
# best_obtained_with_inversions = amélioration par inversion pour améliorer les tours obtenus par les deux précédentes méthodes.

#Lorsqu'il s'agit de circuits de quatre villes, l'exécution de tour_from_closest_city est plus rapide que celle de best_tour_from_closest_city. Cependant, 
#en contrepartie, les circuits générés par tour_from_closest_city ne sont pas toujours précis. 
#Bien qu'ils nécessitent moins de temps, ils présentent souvent une moindre précision. 
#C'est là que l'algorithme best_obtained_with_inversions intervient : il améliore les circuits créés par tour_from_closest_city 
#en introduisant des inversions. Toutefois, cette amélioration se fait au prix 
#d'une augmentation significative du temps d'exécution par rapport aux deux autres fonctions mais toujours donne le plus petit tours.

#Cependant, avec l'augmentation du nombre de villes dans le tableau, les résultats des deux fonctions, 
#best_tour_from_closest_city et tour_from_closest_city, ne garantissent pas nécessairement le circuit le plus court. 
#En particulier, la fonction tour_from_closest_city produit des circuits qui deviennent carrément faux à mesure que le nombre de villes augmente. 
#L'impact des erreurs augmente significativement dans cette fonction, 
#avec best_obtained_with_inversions fournissant beaucoup  d'améliorations sur les circuits générés.

#Pour best_tour_from_closest_city, bien qu'elle présente également des erreurs croissantes avec un nombre croissant de villes,
#elle maintient une meilleure précision globale que tour_from_closest_city. Malgré cela,
#best_tour_from_closest_city reste moins précise que best_obtained_with_inversions. Cependant, il est essentiel de noter que,
#malgré son manque de précision accrue, tour_from_closest_city demeure plus rapide que best_tour_from_closest_city. 
#Ainsi, le choix entre les deux dépend du compromis souhaité entre vitesse d'exécution et précision du résultat en fonction du nombre de villes dans le tableau.
```
